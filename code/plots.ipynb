```python
import os
import pickle
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
dataPath = '/project/zunderlab/heussner/Projects/morph/data/CODEX_LN/'
savePath = '/project/zunderlab/heussner/Projects/morph/results/'
marker = 'DAPI'
tilePath = os.path.join(dataPath,f'CODEX_LN_tiles/{marker}.pkl')
with open(tilePath,'rb') as handle:
    tiles = pickle.load(handle)
```

```python
example = tiles[482].copy()
```

```python
# create the x and y coordinate arrays (here we just use pixel indices)
xx, yy = np.mgrid[0:example.shape[0], 0:example.shape[1]]

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(elev=70, azim=45, roll=10)
# Create surface plot using pixel intensities as Z values
surface = ax.plot_surface(xx, yy, example, cmap='bwr',vmin=0, vmax=np.max(example), 
                         alpha=0.9, linewidth=0, antialiased=True)
ax.set_xlabel('Y')
ax.set_xticks([])
ax.set_ylabel('X')
ax.set_yticks([])
ax.set_zlabel('Probability')
plt.tight_layout()
plt.savefig(os.path.join(savePath,'3D_surfacePlot.png'),dpi=500,bbox_inches='tight')
plt.show()
```

```python
cbar = plt.imshow(tiles[482],cmap='bwr')
plt.colorbar(cbar,label='Probability')
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.savefig(os.path.join(savePath,'2D_Plot.png'),dpi=500,bbox_inches='tight')
plt.show()
```
